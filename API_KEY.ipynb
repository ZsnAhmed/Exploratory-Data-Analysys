```python
import requests
import pandas as pd
from prettytable import PrettyTable
import json
```

```python
table_obj = PrettyTable()
Key_val = '09d9425c-0ae4-4eca-8489-ac7e9e2c4569'
```

```python
api_endpoint = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest?CMC_PRO_API_KEY='
api_endpoint += Key_val
api_endpoint
```

```python
json_data = requests.get(api_endpoint).json()
json_data
```

```python
requests.get(api_endpoint).json()
```

```python
cypto = json_data['data']
```

```python
for currency in cypto:
    curr_name =currency['name']
    curr_price = currency['quote']['USD']['price']
    curr_change_1h = currency['quote']['USD']['percent_change_1h']
    curr_change_24h = currency['quote']['USD']['percent_change_24h']
    curr_change_7d = currency['quote']['USD']['percent_change_7d']
    curr_change_30d = currency['quote']['USD']['percent_change_30d']
    table_obj.add_row([curr_name,curr_price,curr_change_1h,curr_change_24h,curr_change_7d,curr_change_30d])
```

```python
table_obj.field_names=['Currency_name','Currency_price','Currency_change_1h','Currency_change_24h','curr_change_7d','curr_change_30d']
```

```python
print(table_obj)
```

```python
table_txt=table_obj.get_string()
with open('output.txt', 'w') as file:
    file.write(table_txt)
```
